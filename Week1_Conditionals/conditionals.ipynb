# Conditionals
- Conditionals allow your program to make decisions, choosing one path over another depending on specified conditions
- Conditionals allow you, the programmer, to allow your program to make decisions: As if your program has the choice between taking the left-hand road or the right-hand road based upon certain conditions.


```python
score = int(input("what's your score? "))
if score > 100 or score < 0: 
    print ("You score is invalid")
elif score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
elif score >= 50:
    print("E")
else:
    print("Fail")
```

# Note:
- score >= 90 and score <100 could be replaced by 90 <= score < 100

```python
#check the parity of number
n = int(input("Enter a number = "))
if n % 2 == 0: print ("Even")
else: print ("Odd")
```

# Short Hand If ... Else
-This is called a conditional expression (sometimes known as a "ternary operator").


```python
n = int(input("Enter a number = "))
print ("Even") if n % 2 == 0 else print ("Odd")
```

# Note

```python
a = 330
b = 330
print("A") if a > b else print("=") if a == b else print("B")
```

# Note
- Ternary operators are particularly useful for simple assignments and return statements.

```python
x = int(input("Enter the value of X = "))
y = int(input("Enter the value of Y = "))
print("X is greater") if x>y else print("Y is greater") if x<y else print("both are equal") 
```

# Match
- Instead of writing many if..else statements, you can use the match statement.
- The match statement selects one of many code blocks to be executed.
- The match expression is evaluated once.
- The value of the expression is compared with the values of each case.
- If there is a match, the associated block of code is executed.

```python
#program to tell you where you live
name = input("What's your name? ")

match name:
    case "Ali":
        print("Islamabad")
    case "Ahmad":
        print("Islambad")
    case "Junaid":
        print("Islamabad")
    case "Imran":
        print("Peshawar")
    case _:
        print("Don't Know")
```

```python
#optimized version
name = input("What's your name? ")

match name:
    case "Ali" | "Ahmad" | "Junaid": # The | operator means OR.

If the name is Ali OR Ahmad OR Junaid,
the same code block executes.
        print("Islamabad")
    case "Imran":
        print("Peshawar")
    case _:
        print("Don't Know")
```

# If Statements as Guards
- You can add if statements in the case evaluation as an extra condition-check:

```python
month = 5
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5 if month == 4:
    print("A weekday in April")
  case 1 | 2 | 3 | 4 | 5 if month == 5:
    print("A weekday in May")
  case _:
    print("No match")
```

# Note
- match doesn't support checking range, but could be done with guard if, but better way would be if else.

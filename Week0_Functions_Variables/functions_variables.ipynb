# Print()
- print(*objects, sep=' ', end='\n', file=None, flush=False)
 (python documentation)
 -*objects means it can take any number of strings
- The function takes an argument called end, and the default is to create a new line 
- print function automatically includes the argument end='\n'
- This \n indicates that the print function will automatically create a line break when run.
- we can provide an argument for the end, so it doesn't create a new line
- That provided value will override the \n


```python
# printing Hello, World in two lines, by using two print
print("Hello, ") #print add an \n at the end and create a new line
print("World") 

# printing Hello, World in one line by using two print
print("Hello, ",end="") #\n has been override
print ("World")

# overriding sep default value
print("Hello, ", "world", sep=":")
```

# Formatting Strings
- An f-string (formatted string literal) is a way to insert variables or expressions directly inside a string.
- the f in print(f"hello, {name}")
- it will evaluate whatever is in the curly braces, like, print the value of the variable or result of some calculations

# Placeholders and Modifiers
- A placeholder can contain variables, operations, functions, and modifiers to format the value
- A placeholder can include a modifier to format the value.
- A modifier is included by adding a colon : followed by a legal formatting type, like .2f which means  fixed point number with 2 decimals:


```python
name = input("what's your name? ")
print(f"Hello, {name}")
x = 5; y = 10 # statements must be separated by newline or semi colon
print(f"{x} + {y} = {x+y}")

# modifier
price = 59
print(f"The price is {price:.2f} dollars")
```

# Strings methods for formatting
- stip(): remove spaces from the string
- capitalize(): capitalize the first letter of the string
- title(): capitalize first letter of each word in the string
- split(): split the name on the basis of embedded spaces
- Upper(): convert it into upper case
- lower(): convert it into lower case

# lenght of the string
- len() returns the number of characters in the strings


```python
name = input("What's your name? ")

#removes spaces from the name (from beginning and end)
name = name.strip() 
print(f"Hello, {name}")

#capitalize the first letter of name
name = name.capitalize()
print(f"Hello, {name}") 

#capitalize first letter of each word of the name
name = name.title()
print(f"Hello, {name}")

#convert it into upper case
print(name.upper())

#conver it into lower case
print(name.lower())

#split the name on the basis of embedded spaces
first, middle, last = name.split(" ")
print(f"Hello, Mr. {first}")



```

# Modify the above code to be more efficient and cleaner

```python
name = input("What's your name? ").strip().title()
first, middle, last = name.split(" ")
print(f"Hello, Mr. {first} {middle} {last}")
```

# round function 
- round() function gives you the nearest integer by default
- round(number[, ndigits]) : if you specify the ndigits, then it will only gives that number of digits after decimal 

```python
num = float(input("Enter a num = "))

print(round(num)) #gives you the nearest integer by default
print(round(num, 2)) #gives you max 2 no of digits after decimal
```

## Function Names
- Function names follow the same rules as variable names in Python:
- A function name must start with a letter or underscore

- Function names are case-sensitive (myFunction and myfunction are different)

# defining hello in the last
- We don’t have to have our function at the start of our program. We can move it down, but we need to tell the interpreter that we have a main function and a separate hello function.
- When Python reads hello(name) inside main(), why doesn't it give an error since hello() is defined later?
- Because Python reads and creates all function definitions first, before it executes main().


```python
# define main, but it will not be executed first
def main():
    name = input("What's your name")
    hello(name) 

def hello(to):
    print("Hello, ", to)

main()
```

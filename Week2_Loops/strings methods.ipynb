# string formatting
- join() method
- title()
- strip()

```python
students = ["ali khan  ", "  amjad abbasi", "   mujeeb ali", "  muqadar", "irfan", "jamshed"]
studentsformatted = []
print(students)
for s in students:
    studentsformatted.append(s.strip().title())

print(studentsformatted)
print(" | ".join(studentsformatted)) 
# join is a string method to take a collection, and place the string in between the items of the collection
```

# string slicing

```python
phone = "617-495-1000"
#print first 3 digits
print(phone[0:3]) #starting inclusive and ending exclusive
#or
print(phone[:3])
# last four digits
print(phone[-4:])
#or
print(phone[8:])
```

# Python Collections (Arrays)
- There are four collection data types in the Python programming language:
## List 
- is a collection which is ordered and changeable. Allows duplicate members.
## Tuple
- is a collection which is ordered and unchangeable. Allows duplicate members.
## Set
- is a collection which is unordered, unchangeable*, and unindexed. No duplicate members.
## Dictionary
- is a collection which is ordered** and changeable. No duplicate members.
## Note
- *Set items are unchangeable, but you can remove and/or add items whenever you like.
- **As of Python version 3.7, dictionaries are ordered. In Python 3.6 and earlier, dictionaries are unordered.

# List Items
- List items are ordered, changeable, and allow duplicate values
- List items are indexed, the first item has index [0], the second item has index [1] etc.
## Ordered
- When we say that lists are ordered, it means that the items have a defined order, and that order will not change.
- If you add new items to a list, the new items will be placed at the end of the list.
- Note: There are some list methods that will change the order, but in general: the order of the items will not change.
## Changeable
- The list is changeable, meaning that we can change, add, and remove items in a list after it has been created.

# Access Items
- List items are indexed and you can access them by referring to the index number:
- The first item has index 0.
## Negative Indexing
- Negative indexing means start from the end
- -1 refers to the last item, -2 refers to the second last item etc.
## Range of Indexes
- You can specify a range of indexes by specifying where to start and where to end the range.
- When specifying a range, the return value will be a new list with the specified items.
# Note
## Python slices use:
- [start:end]
- where: start is included ✅ and end is excluded ❌
- This design makes slicing and range() work consistently and allows slices to fit together without overlaps or gaps
# Range of Negative Indexes
- Specify negative indexes if you want to start the search from the end of the list:



```python
mylist = ["item1", "item2", "item3", "item4", "item5", "item6"]
print(mylist[1:4]) #start at index 1 (included) and end at index 5 (not included).
print(mylist[1:]) #print from index 1 to end (including end)
print(mylist[:3]) #print from index 0 to 2 (exluding end)
# This example returns the items from "item2" (-5) to "item4" (-3), but NOT including "item5" (-2): 
# as end is excluded
# slicing move from left to right
print(mylist[-5:-2])
# slicing move from left to right
print(mylist[-2:-5]) #print nothings
# wanna move backwards, use step
print(mylist[-2:-5:-1])
```

# Check if Item Exists
- To determine if a specified item is present in a list use the in keyword:

```python
mylist = ["item1", "item2", "item3", "item4", "item5", "item6"]
if "item8" in mylist:
    print("item8 Found")
else:
    print("item8 Not Found")
if "item3" in mylist:
    print("item3 Found")
else:
    print("item3 Not Found")
```

# list methods
- 

```python
mylist = ["item1", "item2", "item3", "item4", "item5", "item6"]

mylist.append("item7") #add the item at the end
print("Added item7: ", mylist)
# want to add the multiple items
mylist.append(["item8", "item9"]) #this will insert a separate list
print("Added a list: ", mylist)
# first remove these items
mylist.remove(["item8", "item9"])
print("removed the added list: ", mylist)
# then add as separate items, use extend method, which will merge the given list with the mylist
mylist.extend(["item8", "item9"])
print("extended the list: ", mylist)
# now just remove a single item
mylist.remove("item3") # it searches for item3 in the list, and removes it
print("Removed item3: ", mylist)
# now if want to insert a item at a specific location, such as at index 4
mylist.insert(4, "item5.5")
print("Inserted item7 at index 4: ", mylist)
# reverse the list
mylist.reverse()
print ("Reversed the list: ", mylist)
```

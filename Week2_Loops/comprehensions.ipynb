# List Comprehension
- List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list
- Without list comprehension you will have to write a for statement with a conditional test inside:
## The Syntax
- newlist = [expression for item in iterable if condition == True]
### Condition
- The condition is like a filter that only accepts the items that evaluate to True.
### Iterable
- The iterable can be any iterable object, like a list, tuple, set etc.
### Expression
- The expression is the current item in the iteration, but it is also the outcome, which you can manipulate before it ends up like a list item in the new list:
- Set the values in the new list to upper case: newlist = [x.upper() for x in fruits]

```python
students = ["ali", "amjad", "akram", "wasim", "moeen", "yonus"]
# lets create a new list from this list, by adding those students, having a in the names
# without comprehension
newlist = []
for x in students:
    if "a" in x:
        newlist.append(x)
print(newlist)
newlist.clear()

print(newlist)
# By comprehension
newlist = [x for x in students if "a" in x] 
# simply write the three lines, i.e. loop, condition and assignment in online
print(newlist)
```

# Dictionary
- A dictionary comprehension is a concise way to create a dictionary, similar to how list comprehensions create lists.
- offers a shorter syntax when you want to create a new dictionary based on the values of an existing list or dictionary
- General Syntax: {key: value for item in iterable}

```python
students = {
    "ali", "amjad", "akbar"
}
nameslength = {s:len(s) for s in students}
print(nameslength)
```

```python
# from 1 to 10, print squares of even only
evensquares = {x:x**2 for x in range(1, 10) if x%2==0}
print(evensquares)
```

```python
# create a spacecraft report
def main():
    spacecraft = {
        "Name":"Ghori 1",
        "Distance": 0.01,
    }
    spacecraft["Orbit"] = "Sun" # add another pair of key and value
    spacecraft.update({"Company":"NASA", "Country": "America"}) # add more key-value pairs 
    print(create_report(spacecraft))

def create_report(spacecraft):
    return f"""
    =========REPORT=========
    {"Name"}: {spacecraft["Name"]} 
    {"Distance"}: {spacecraft["Distance"]} AU
    {"Orbit"}: {spacecraft["Orbit"]}
    {"Company"}: {spacecraft["Company"]}
    Country: {spacecraft.get("Country", "Unknown")} 
    {"City"}:  {spacecraft.get("City", "Unknown")}
    =========================
    """
# spacecraft.get() if you wanna some default value

main()
```

```python
students = {
    "Name":"Ali",
    "School":"ICB",
    "City": "Islamabad"
}
for student in students: #for student in students.keys() will return the same, i.e. the keys in student
    print(f"{student}: {students[student]}")
for student in students.values(): #this will return the values in student
    print(f"{student}")
print(students.items()) # items return key and values both; so, it will print all the key and values
```

```python
# spelling bee
words = {
    "pair":4,
    "hair":4,
    "chair":5,
    "graphic":7
}

def main():
    print("Spelling Bee: ")
    print("Your letters are A I P C R H G")

    while len(words) > 0:
        print(f"{len(words)} left")
        guess = input("Guess a word? ")
        if guess == "graphic":
            print("You won")
            for word, points in words.items(): # key, value both will be returned by words.items()
                print(f"{word} has {points}") 
            words.clear() #delete entire dictionary
            # break can also be used 
        if guess in words.keys():
            points = words.pop(guess) # it will delete the guessed word and its points, and return the points
            print(f"Good! you scored {points}")
main()
 
```

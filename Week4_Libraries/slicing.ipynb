# slice
- slice is a command that allows us to take a list and tell the interpreter where we want the interpreter to consider the start of the list and the end of the list

## Python slicing syntax is:

list[start:end:step]

To slice **backwards**, use a **negative step**.

# 1. Reverse the Entire List

mylist = [10, 20, 30, 40, 50]
print(mylist[::-1])

Output: [50, 40, 30, 20, 10]

Meaning: [start:end:-1]

Since `start` and `end` are omitted, Python starts from the last element and moves backward.

# 2. Slice a Portion in Reverse

mylist = [10, 20, 30, 40, 50]

print(mylist[4:1:-1])
Output: [50, 40, 30]

Start at index 4  → 50
Move backward
Stop before index 1 (end is omitted)

# 3. Using Negative Indices

mylist = [10, 20, 30, 40, 50]

print(mylist[-1:-4:-1])
Output: [50, 40, 30]

Stop before -4


# 4. Last Three Elements (Reversed)


print(mylist[:-4:-1])

Output: [50, 40, 30]

# 5. Every Second Element (Reversed)

print(mylist[::-2])

Output: [50, 30, 10]

# Visualization

List:
Value : 10  20  30  40  50
Index :  0   1   2   3   4
Neg   : -5  -4  -3  -2  -1



```python
mylist = [10, 20, 30, 40, 50]

print(mylist[3:0:-1]) # 40 30 20 
print(mylist[-2::-1]) # 40 30 20 10
print(mylist[:1:-1]) # 50 40 30
```

```python
mylist = [10, 20, 30, 40, 50]

print(mylist[-1:-5:-2]) # 50 30
print(mylist[2::-1]) # 30 20 10
print(mylist[-3:0:-1]) # 30 20 
print(mylist[-4:-1]) # 20 30 40  
print(mylist[-1::-2]) # 50 30 10
```

# Python Style (PEP 8)

## What is Style?
- Style refers to how code is written and formatted.
- Good style makes code easier to read, understand, and maintain.
- Readable code reduces the chances of introducing bugs.

## PEP 8
- PEP = Python Enhancement Proposal.
- PEP 8 is the official Python style guide.
- It provides conventions for writing clean and consistent Python code.
- The main principle of PEP 8 is:
    - Readability counts

## Style Guide
- A style guide defines rules for formatting code.
- Different companies or projects may have their own style guides.
- Consistency is more important than personal preference.

## Common PEP 8 Guidelines

- Use 4 spaces for indentation (do not use tabs).
- Keep lines reasonably short.
- Use blank lines to separate logical sections of code.
- Organize import statements at the beginning of the file.
- Write consistent and readable code.

## Linters
- A linter analyzes source code for style issues and possible mistakes.
- It does not execute the program.
- Example:
  - pylint

## Code Formatter
- A code formatter automatically reformats code according to a style guide.
- It improves readability without changing the program's behavior.

Examples:
- pycodestyle
- black

## Black

- Black is an opinionated Python code formatter.
- It automatically formats code according to its own rules.
- It follows PEP 8 conventions with very few configuration options.
- Purpose: Spend less time formatting code and more time solving problems.
- python -m pip install black
- python -m black filename.py

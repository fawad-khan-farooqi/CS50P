# libraries
- Generally, libraries are bits of code written by you or others that you can use in your program.
- Python allows you to share functions or features with others as “modules”
- It’s easier as a coder to stand on the shoulders of prior coders.
# import
- you can load a module into your program, by using the keyword "import" followed by the module name

# random
- random is a library that comes with Python that you could import into your own project
- you can access a function defined in random module by using the module name . and the name of function
- random.choice()

# from
- from allows us to be very specific about what we’d like to import. 
- Prior, our import line of code is bringing the entire contents of the functions of random. 
- However, what if we want to only load a small part of a module?
- from random import choice # now can import just the choice function of random
- now, you can directly access choice function without mentioning the module name with it
- choice()

# few functions of random module
- random.choice(list) # return one value from the list randomly
- random.randint(1,10) # return one value from 1 to 10 (including 1 and 10)
- random.shuffle(list) # shuffle the list items in place unlike the other functions

# statistics
- a module, comes with Python, containing statistics functions like, mean, mode, median etc.
- statistics.mean(listofvalues) # return the average of the list of values  


```python
import random #import random module (a library, containing built-in functions, etc.)
# it comes with python installation, but need to be imported separately

choice = random.choice(["Head", "Tail"])
print(choice)
```

```python
# if you want to only import choice from random module 
# and don't want to write random with choice everytime

from random import choice # from random module, only choice is imported

print(choice(["head","tail"]))
```

```python
# random.randint()
import random

print(random.randint(1, 10)) # will return a number from 1 to 10, randomly, inclusively (1 and 10)
```

```python
# random.shuffle()
import random

cards = ["queen", "king", "jack"]

random.shuffle(cards) # this will shuffle the cards in place, unlikely, other functions

for card in cards:
    print(card)
```

```python
# import statistic
import statistics

marks = [80, 90]
print(statistics.mean(marks))
```

# Making Your Own Python Libraries

- You can create your own Python module to reuse code.
- A Module is simply a Python file (`.py`) containing functions, classes, or variables.
- Creating your own modules avoids code duplication and makes programs easier to maintain.
- Save your reusable code in a Python file (e.g., `calculator.py`).
- Create another Python file in the same directory and import the module.

Example:

import calculator

calculator.add(5, 3)
- After importing, you can directly use the functions, classes, or variables defined in the module.

## if __name__ == "__main__":
    main()

at the end of the module when you want main() to run only if the file is executed directly.

- If the file is imported into another program, main() will not execute automatically.

This allows the same file to be used both as:
1. A standalone program.
2. A reusable module

## Module vs package vs library

- Module: A single Python file (.py).
- Package: a folder containing one or more modules (and possibly subpackages).
- Library: A reusable collection of modules/packages
    - A library may consist of:
        - one module, one package, or many packages and modules.

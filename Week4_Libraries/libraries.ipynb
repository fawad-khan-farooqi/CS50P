```python
import random #import random module (a library, containing built-in functions, etc.)
# it comes with python installation, but need to be imported separately

choice = random.choice(["Head", "Tail"])
print(choice)
```

```python
# if you want to only import choice from random module 
# and don't want to write random with choice everytime

from random import choice # from random module, only choice is imported

print(choice(["head","tail"]))
```

```python
# random.randint()
import random

print(random.randint(1, 10)) # will return a number from 1 to 10, randomly, inclusively (1 and 10)
```

```python
# random.shuffle()
import random

cards = ["queen", "king", "jack"]

random.shuffle(cards) # this will shuffle the cards in place, unlikely, other functions

for card in cards:
    print(card)
```

```python
# import statistic
import statistics

marks = [80, 90]
print(statistics.mean(marks))
```

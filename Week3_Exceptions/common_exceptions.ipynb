# CS50P, NumPy, Pandas, AI Engineering, and everyday Python, you only need to remember about 10 exceptions initially.

## 1. ValueError
- Raised when the type is correct but the value is invalid.
E.g: int("abc")
Output: ValueError: invalid literal for int()
- Used constantly in CS50P.

## 2. NameError 
- Raised when a variable doesn't exist.
e.g: print(x)
Output: NameError: name 'x' is not defined

## 3. TypeError 
- Raised when an operation uses the wrong type.
e.g: "Hello" + 5
Output: TypeError

## 4. IndexError
- Raised when a list index doesn't exist.
e.g: mylist = [1, 2, 3]
     print(mylist[5])
Output: IndexError
## 5. KeyError 
- Raised when a dictionary key doesn't exist.
e.g: student = {"name": "Fawad"}
    print(student["age"])
Output: KeyError

## 6. FileNotFoundError 
- Raised when a file cannot be found.
e.g:  open("data.txt")
Output: FileNotFoundError
- Very common in File I/O.

## 7. ZeroDivisionError 
- Raised when dividing by zero.
e.g: 10 / 0
Output: ZeroDivisionError

## 8. EOFError 
- Raised when input unexpectedly ends.
e.g: try:
        item = input()
     except EOFError:
        break

## 9. AttributeError
- Raised when an object doesn't have a method or attribute.
e.g: mylist = []
     mylist.items()
Output: AttributeError: 'list' object has no attribute 'items'

## 10. ModuleNotFoundError
- Raised when importing a module that isn't installed.
e.g: import numpyy
Output: ModuleNotFoundError
- Very common when learning libraries.

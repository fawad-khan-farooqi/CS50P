# Exceptions
- Exceptions are problems that arise while your program is running.

# Runtime Errors
- Runtime errors refer to those created by unexpected behavior within your code.
- For example, perhaps you intended for a user to input a number, but they input a character instead. 
- Your program may throw an error because of this unexpected input from the user.
- As programmers, we should be defensive to ensure that our users are entering what we expected.
- If we run this program and type “cat”, we’ll see ValueError: invalid literal for int() with base 10: 'cat'. 
- In other words, the int function cannot convert the text “cat” into a number.
- An effective strategy to fix this potential error would be to create “error handling” to ensure the user behaves as we intend.

# Exception Handling
- When an error occurs, or exception as we call it, Python will normally stop and generate an error message.
- These exceptions can be handled using the try statement

# Python Try Except
- In Python try and except are ways of testing out user input before something goes wrong
- The try block lets you test a block of code for errors.
- The except block lets you handle the error.
- The else block lets you execute code when there is no error.
- The finally block lets you execute code, regardless of the result of the try- and except blocks.


```python
# The try block will generate an exception, because x is not defined:
try:
  print(x)
except:
  print("An exception occurred")
# Since the try block raises an error, the except block will be executed.
# Without the try block, the program will crash and raise an error:
```

# Many Exceptions
- You can define as many exception blocks as you want,
 - e.g. if you want to execute a special block of code for a special kind of error

```python
# Print one message if the try block raises a NameError and another for other errors
try:
  print(x)
except NameError:
  print("Variable x is not defined")
except:
  print("Something else went wrong")
```

# Else
- You can use the else keyword to define a block of code to be executed if no errors were raised
# Finally
- The finally block, if specified, will be executed regardless if the try block raises an error or not.

```python
try:
    print(x)
except:
    print("This block will be executed only if rrror found in the try block")
else:
    print("This block will be executed only if no error found in the try block")
finally:
  print("This block will be executed regardless of error found or not")
```

# Raise an exception
- You can throw (or raise) an exception if a condition occurs
    - by using the raise keyword

```python
# Raise an exception, if x is negative
x = -1
if x < 0:
  raise Exception("Sorry, no numbers below zero")
```

- You can define what kind of error to raise, and the text to print to the user.

```python
# Raise a TypeError if x is not an integer
x = "hello"

if not type(x) is int:
  raise TypeError("Only integers are allowed")

print("Program continues...")
```

# Note:
- In the above example, you can simply use print to display the message
- But, the main differences are
## if  
- A message was displayed. (inform the user)
- The program continued running
- Python did not consider it an error.
## raise
- An exception was raised. (signal an error to Python or another programmer.)
- The program stopped immediately.
The line: print("Program continues...") never executes

## try-except and raise: Key Points
- When an exception occurs inside a try block, Python immediately skips the remaining statements in the try block and jumps to the matching except block.
- Raising an exception does not necessarily stop the program.
- If an exception is raised and a matching except block catches it, the program can continue execution.
- If an exception is raised and no matching except block catches it, the program stops and displays an error message.
- Raising an exception outside a try-except block usually stops the program.
- An exception can be re-raised inside an except block using raise, which usually causes the program to stop unless another try-except block catches it.
- The purpose of try-except is to handle errors gracefully instead of crashing the program.


```python
# Example
try:
    raise ValueError("Oops")
except ValueError:
    print("Handling...")
    raise
```
